```python
import pandas as pd
```

```python
from sklearn import linear_model
```

```python
df=pd.read_csv("dataq5.csv")
```

```python
df
```

```python
X = df[['Weight', 'Volume']]
Y = df['CO2']
```

```python
X
```

```python
Y
```

```python
regr = linear_model.LinearRegression()
```

```python
regr.fit(X, Y)
```

```python
predictedCO2 = regr.predict([[2300, 1300]])
```

```python
print(predictedCO2)
```
